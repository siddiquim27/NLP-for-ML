```python
corpus = ''' hello and welcome , tp noman siddiquis world. please mujhe koi genAi sikha do'''
```

```python
corpus
```

```python
## tokenization
## sentence to para
```

```python
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
documents
```

```python
## tokenization
## para to words
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)  ## make sure that every punctuation also separates
```

```python
from nltk.tokenize import TreebankWordTokenizer  ## isko use kiye to full stop akhiri word k sath ayga
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
