```python
words = ["eating", "eats", "eaten", "write", "writing", "written", "programs", "programming"]
```

```python
words
```

```python
## PORTER STEMMER

from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
for word in words:
  print(word+ "------> "+stemming.stem(word))
```

```python
stemming.stem("congratukations") ## yha congratulations ki meaning hi bdl di isne
```

```python
stemming.stem("unnecessary")
```

```python
## Regex Stemmer class

## dollar ka sign last me rha to wo last ka remove krega
## aur dollar ka sign phle rha to wo phle ka remove krega
```

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("ingeating")
```

```python
from nltk.stem import SnowballStemmer
snow_stemmer = SnowballStemmer(language='english')
```

```python
for word in words:
  print(word + "------->" + snow_stemmer.stem(word))
```
