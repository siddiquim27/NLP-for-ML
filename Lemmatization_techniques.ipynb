```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')  # Optional, improves lemmatization

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize('cars'))
```

Lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('cars' ,pos = 'v')
```

```python
lemmatizer.lemmatize('boxes', 'r')
```

```python
lemmatizer.lemmatize('geese', 'a')
```

```python
lemmatizer.lemmatize('children', 'n')
```

```python
lemmatizer.lemmatize('fairly')
```
